```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('parrot.jpeg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img_crop = img[400:, :80]


ax, fig = plt.subplots(1, 2)
fig[0].imshow(img)
fig[0].title.set_text('ORIGINAL')

fig[1].imshow(img_crop)
fig[1].title.set_text('CROP')

plt.show()
```

```python
height, width = img.shape[:2] # From original photo

print("Height: ", height)
print("Width: ", width)

height_c, width_c = img_crop.shape[:2] # Frame that we have to find

print("Height_crop: ", height_c)
print("Width_crop: ", width_c)


min_diff = float('inf')         #infinity
for i in range(0, height - height_c):
    for j in range(0, width - width_c):
        img_check = img[i:i+height_c, j:j+width_c]
        difference = np.sum(np.abs(img_check - img_crop))
        if difference < min_diff:
            min_diff = difference
            best_i = i
            best_j = j

box = cv.rectangle(img, (best_j, best_i), (best_j+width_c, best_i+height_c), (255,255,0), 3)
plt.imshow(box)
plt.show()
```
