```python
import cv2 as cv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
templateImagePath="bird_template.jpg"
mainImagePath="birds.jpg"

templateImg=cv.imread(templateImagePath,0)
mainImg = cv.imread(mainImagePath,0)
templatewidth , templateHight = templateImg.shape
```

```python
methods = ['cv.TM_CCOEFF', 'cv.TM_CCOEFF_NORMED', 'cv.TM_CCORR',
 'cv.TM_CCORR_NORMED', 'cv.TM_SQDIFF', 'cv.TM_SQDIFF_NORMED']
```

```python
result = cv.matchTemplate(mainImg,templateImg,cv.TM_CCOEFF_NORMED)
plt.imshow(result, cmap = 'gray')
```

```python
minIntensity,maxIntensity,minlocation,maxlocation = cv.minMaxLoc(result)
```

To select multipple objects 

```python
y_prev =0
threshold = 0.5
loc = np.where(result >= threshold)

reoderedList = zip(*loc[:: -1])

for (x,y) in reoderedList:
    if y-y_prev >=20:
        cv.rectangle(mainImg, (x,y), (x + templateHight, y + templatewidth), (255, 0, 0), 2)
        y_prev = y

cv.imshow('img', mainImg)
cv.waitKey(0)
cv.destroyAllWindows()
```
